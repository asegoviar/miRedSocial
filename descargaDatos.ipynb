```python
import os
import django

os.environ.setdefault("DJANGO_SETTINGS_MODULE", "miRedSocial.settings")
django.setup()
```

```python
from asgiref.sync import sync_to_async
from django.contrib.auth.models import User

async def obtener_usuarios():
    return [u async for u in User.objects.all()]

usuarios = await obtener_usuarios()

for u in usuarios:
    print("Hola ->", u.username)
```

```python
from asgiref.sync import sync_to_async
from django.contrib.auth.models import User
import pandas as pd

# ðŸš€ Convertimos la consulta en una lista sÃ­ncrona usando async wrapper
usuarios = await sync_to_async(list)(User.objects.all())

data = []

for u in usuarios:
    data.append({
        "id": u.id,
        "username": u.username,
        "email": u.email,
        "first_name": u.first_name,
        "last_name": u.last_name,
        "password_hash": u.password,
        "is_active": u.is_active,
        "is_staff": u.is_staff,
        "is_superuser": u.is_superuser,
    })

df = pd.DataFrame(data)  
df
```

```python
df.to_csv("usuarios_exportados.csv", index=False)
print("CSV generado correctamente ðŸš€")
```

```python
from asgiref.sync import sync_to_async
from accounts.models import Profile
import pandas as pd

# ðŸš€ Convertimos el queryset a lista sincronizable
perfiles = await sync_to_async(list)(Profile.objects.select_related("user").all())

data = []

for p in perfiles:
    data.append({
        "profile_id": p.id,
        "user_id": p.user.id,          # ðŸ”¥ REFERENCIA AL USER
        "username": p.user.username,   # ðŸ§  Nombre del usuario relacionado
        "bio": p.bio,
        "birthdate": p.birthdate,
        "is_private": p.is_private,
        "created_at": p.created_at,
        "updated_at": p.updated_at,
    })

df_profiles = pd.DataFrame(data)
df_profiles
```

```python
df_profiles.to_csv("profiles_exportados.csv", index=False)
print("âœ” CSV generado: profiles_exportados.csv")
```
